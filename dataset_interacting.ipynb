```python
# Lọc sinh viên năm nhất
firstYear_df = ungraduated_df[ungraduated_df['diemtbhk_2'] == 0]

# Lọc sinh viên năm hai
secondYear_df = ungraduated_df[(ungraduated_df['diemtbhk_4'] == 0) & (ungraduated_df['diemtbhk_2'] != 0)]

# Lọc sinh viên năm ba
thirdYear_df = ungraduated_df[(ungraduated_df['diemtbhk_6'] == 0) & (ungraduated_df['diemtbhk_4'] != 0)]

# Lọc sinh viên năm tư
fourthYear_df = ungraduated_df[(ungraduated_df['diemtbhk_8'] == 0) & (ungraduated_df['diemtbhk_6'] != 0)]
```
